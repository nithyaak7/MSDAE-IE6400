Using this Jupyter Notebook, practice along with the three tasks listed below:
- Generate a sample dataset using make_blobs.​
- Cluster the dataset into 4 groups using K-Means.​
- Plot the clusters with their respective centroids.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generating random data for clustering
data, _ = make_blobs(n_samples=300, centers=4, random_state=42)

# Using K-Means for clustering
kmeans = KMeans(n_clusters=4)
clusters = kmeans.fit_predict(data)

# Plotting the clusters and their centers
plt.scatter(data[:, 0], data[:, 1], c=clusters, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='black', marker='X')
plt.title("K-means Clustering with 4 Clusters")
plt.show()
```

Next, use the same dataset from the K-means exercise.  
Apply Agglomerative Clustering for hierarchical clustering.   
Visualize the 4 distinct clusters using a scatter plot.

```python
from sklearn.cluster import AgglomerativeClustering

# Using Agglomerative Clustering for hierarchical clustering
agg_clustering = AgglomerativeClustering(n_clusters=4, linkage='ward')
agg_clusters = agg_clustering.fit_predict(data)

# Plotting the clusters
plt.scatter(data[:, 0], data[:, 1], c=agg_clusters, cmap='rainbow')
plt.title("Hierarchical Clustering with 4 Clusters")
plt.show()
```

Use the linkage method to determine the hierarchical order.  
Generate a dendrogram to visualize the hierarchy.  
Observe the tree structure and merging clusters.  

```python
from scipy.cluster.hierarchy import dendrogram, linkage

# Generating a dendrogram
linked = linkage(data, 'ward')

plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top')
plt.title("Dendrogram for Hierarchical Clustering")
plt.show()
```
