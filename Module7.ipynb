#### Applying ACF

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from statsmodels.tsa.stattools import acf
from matplotlib.ticker import MaxNLocator


# Create a time series with a trend
np.random.seed(0)
time = pd.date_range(start='2000-01-01', periods=100, freq='M')
trend = np.linspace(0, 10, 100)
noise = np.random.normal(0, 1, 100)
time_series = trend + noise

# Calculate ACF
acf_values = acf(time_series, nlags=20)
# Create the plot
fig, axs = plt.subplots(2, 1, figsize=(10, 8))
# Time Series Plot
axs[0].plot(time, time_series, label='Time Series with Trend')
# ACF Plot
axs[1].stem(range(len(acf_values)), acf_values, use_line_collection=True)
plt.tight_layout()
# Adjust DPI for high quality
plt.show()
```

#### Applying Multiplicative Decomposition

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose


# Load the flight data
flights = sns.load_dataset("flights")
ts = flights['passengers']

# Perform multiplicative decomposition
result = seasonal_decompose(ts, model="multiplicative", period = 12)

# Create subplots for each component
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(10, 10))
fig.dpi = 300  # Set the figure DPI to 300

# Plot the original data
ax1.set_title("Original Data")
ts.plot(ax=ax1)

# Plot the trend component
ax2.set_title("Trend Component")
result.trend.plot(ax=ax2)

# Plot the seasonal component
ax3.set_title("Seasonal Component")
result.seasonal.plot(ax=ax3)

# Plot the residual component
ax4.set_title("Residual Component")
result.resid.plot(ax=ax4)

plt.tight_layout()
plt.show()
```
